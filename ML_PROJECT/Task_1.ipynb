### Detect Outliers

```python
#Method to handle outliers: IQR,Boxplot
import seaborn as sns
import matplotlib.pyplot as plt

numeric_value=['age','height','weight','ap_hi','ap_lo','cholesterol','gluc','smoke','alco','active','cardio']


for col in numeric_value:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_limit) | (df[col] > upper_limit)]
    print(f"{col} → Outliers found: {outliers.shape[0]}")
```

### Remove Outliers

```python
for col in numeric_value:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
df = df[(df[col] >= lower_limit) & (df[col] <= upper_limit)]
```

### scale numeric features

```python
# perform min-mix scaller so we can convert our data into 0 to 1 range
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

numeric_value=['age','height','weight','ap_hi','ap_lo','cholesterol','gluc','smoke','alco','active','cardio']

df[numeric_value] = scaler.fit_transform(df[numeric_value])
df[numeric_value]
```
