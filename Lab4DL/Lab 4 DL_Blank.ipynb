## Simple Linear Regression

## Step 1. Import the necessary libraries

```python
import pandas as pd
```

## Step 2. Import the dataset

```python
df = pd.read_csv('50_Startups.csv')
```

```python
df
```

## Step 3 . Check the State Column

```python
df['State'].value_counts()
```

## Step 4 . Splitting dataset in to input and output

```python
x = df.iloc[::,:4:]
y = df.iloc[::,4::]
x

#or
#x = df.drop(['Profit'] , axis = 1)
```

## Step 5 . Convert state Column into Numeric Column

## Step 5.1 .  Perform Transformation

```python
x1 = pd.get_dummies(x,columns=['State'],drop_first=True)
# here drop_first =  True it means ke koi state ma 3 column ma divide thai to ema thi koi ek column drop thy jase..!!
# by default biji 2 columns uper thi aapne 3rd column ne predict kri shishu
```

```python
x1
```

## Step 6 . Dummy variable trap

```python
# Already Performed using | drop_first =True
```

## Step 7 Splitting dataset in to Train and Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x1,y,test_size=0.2,random_state=42)
```

## Step 8  Import LinearRegression model from linear_model family 

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

## Step 9  Fit the data

```python
model.fit(x_train,y_train)
```

## Step 10  Predict the data

```python
y_predict = model.predict(x_test)
```

## Step 11  Display Result

```python
# y_test and y_predict
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_test['Administration'],y_test)
plt.scatter(x_test['Administration'],y_predict)
```

## RSS

```python
#RSS means Residual Sum Of Square
#Formula of RSS : [sum of ((yi-yi')^2)]
import numpy as np
```

```python
rss = np.sum((y_test.values - y_predict) ** 2)
rss
```

```python
len(y_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_predict) * len(y_test)
```

```python
len(y_predict)
```

## R Square

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_predict)
```

## Now use Polynomial Regression on Position_Salaries dataset

```python
import pandas as pd
```

```python
data = pd.read_csv('Position_Salaries.csv')
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data['Level'],data['Salary'])
```

```python
x = data[['Level']]
y = data[['Salary']]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
x_train
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
x1 = poly.fit_transform(x_train)
```

```python
x1
```

```python
model1 = LinearRegression()
```

```python
model1.fit(x1,y_train)
```

```python
y_poly_predict = model1.predict(poly.fit_transform(x_test))
y_poly_predict
```
